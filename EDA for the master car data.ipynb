### Total Car Sales by Brand

The bar plot above displays the total car sales (in Lakh) for each car brand. Hyundai leads with the highest total sales, followed closely by Maruti and Renault. The visualization highlights the dominance of these brands in the market, while brands like Toyota and Ford have comparatively lower total sales. This distribution provides insights into brand popularity and market share within the dataset.

```python
sales_by_brand = df.groupby('Brand')['Sale_Price_Lakh'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=sales_by_brand.index, y=sales_by_brand.values, palette='viridis')
plt.xlabel('Brand')
plt.ylabel('Total Sales (Lakh)')
plt.title('Total Car Sales by Brand')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Total Car Sales Trend by Year

The line plot above illustrates the trend of total car sales (in Lakh) over the years. The data shows fluctuations in sales, with notable peaks and troughs. The most recent year, 2024, records a significant increase in total sales, reaching approximately 1671.8 Lakh. This suggests a strong upward trend in car sales for the latest year in the dataset.

```python
sales_by_year = df.groupby('Year')['Sale_Price_Lakh'].sum().sort_index()
plt.figure(figsize=(10,6))
sns.lineplot(x=sales_by_year.index, y=sales_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Total Sales (Lakh)')
plt.title('Total Car Sales Trend by Year')
plt.tight_layout()
# for x, y in zip(sales_by_year.index, sales_by_year.values):
#     plt.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=9)
plt.show()
```

### Distribution of Cars by Fuel Type

The bar plot above illustrates the distribution of cars based on their fuel type. Electric and Petrol vehicles have the highest representation in the dataset, followed closely by CNG, Hybrid, and Diesel. This distribution provides insights into the popularity and adoption of different fuel types among cars in the dataset.

```python
fuel_counts = df['Fuel_Type'].value_counts()
plt.figure(figsize=(8,5))
sns.barplot(x=fuel_counts.index, y=fuel_counts.values, palette='Set2')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.title('Distribution of Cars by Fuel Type')
plt.tight_layout()
plt.show()
```

### Distribution of Cars by Owner Type

The bar plot above shows the distribution of cars based on their owner type. "First" owner cars are the most common, followed by "Second", "Third", and "Fourth" owner types. This indicates that a significant portion of cars in the dataset are still with their original owners, while a considerable number have changed hands multiple times.

```python
owner_type_counts = df['Owner_Type'].value_counts()
plt.figure(figsize=(8,5))
sns.barplot(x=owner_type_counts.index, y=owner_type_counts.values, palette='pastel')
plt.xlabel('Owner Type')
plt.ylabel('Number of Cars')
plt.title('Distribution of Cars by Owner Type')
plt.tight_layout()
plt.show()
```

### Total Car Sales by City

The bar plot above shows the total car sales (in Lakh) for each city. Lucknow leads with the highest total sales, followed by Chennai, Mumbai, and Hyderabad. This visualization highlights the regional distribution of car sales, indicating that these cities have a significant share of the market. The data provides valuable insights into the geographic trends and preferences of car buyers across different cities.

```python
sales_by_city = df.groupby('City')['Sale_Price_Lakh'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=sales_by_city.index, y=sales_by_city.values, palette='coolwarm')
plt.xlabel('City')
plt.ylabel('Total Sales (Lakh)')
plt.title('Total Car Sales by City')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Total Service Cost by Service Type

The bar plot above displays the total service cost for each service type. "Brake Service" incurs the highest total cost, followed by "Oil Change" and "Full Service". This visualization highlights which service types contribute most to overall service expenses, providing valuable insights for cost management and resource allocation in car maintenance.

```python
service_cost_by_type = df.groupby('Service_Type')['Service_Cost'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=service_cost_by_type.index, y=service_cost_by_type.values, palette='mako')
plt.xlabel('Service Type')
plt.ylabel('Total Service Cost')
plt.title('Total Service Cost by Service Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Top 10 Service Centers by Total Service Cost

The bar plot above highlights the top 10 service centers based on their total service cost. CarZ leads with the highest total service cost, followed by Mechanix, FastFix, and Pitstop. This visualization provides insights into which service centers handle the most expensive or highest volume of services, helping identify key players in the car maintenance market.

```python
top_service_centers = df.groupby('Service_Center')['Service_Cost'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_service_centers.index, y=top_service_centers.values, palette='crest')
plt.xlabel('Service Center')
plt.ylabel('Total Service Cost')
plt.title('Top 10 Service Centers by Total Service Cost')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
